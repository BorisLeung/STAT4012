```python
import pandas as pd
genres = pd.read_csv('/kaggle/input/letterboxd/genres.csv')
movies = pd.read_csv('/kaggle/input/letterboxd/movies.csv')
```

```python
import os
from PIL import Image
import pandas as pd

# Assuming you have the DataFrame 'df' with the columns 'id', 'name', 'date', 'tagline', 'description', 'minute', 'rating', 'genre'
kaggle_path = '/kaggle/input/letterboxd/posters/'

# Initialize a list to store the indices of the rows to be dropped
rows_to_drop = []

# Check if the IDs in the DataFrame match the image files and have three channels
for idx, row in movie_info_df.iterrows():
    file_path = os.path.join(kaggle_path, f"{row['id']}.jpg")
    if os.path.exists(file_path):
        try:
            img = Image.open(file_path)
            if img.mode != 'RGB':
                print(f"Invalid image for ID {row['id']}: {row['name']}")
                rows_to_drop.append(idx)
        except:
            print(f"Error opening image for ID {row['id']}: {row['name']}")
            rows_to_drop.append(idx)
    else:
        print(f"Missing image for ID {row['id']}: {row['name']}")
        rows_to_drop.append(idx)

# Drop the rows with missing or invalid images
movie_info_df = movie_info_df.drop(movie_info_df.index[rows_to_drop])

# Reset the index of the DataFrame
movie_info_df = movie_info_df.reset_index(drop=True)
```

Create the MoviePosterDataset

```python
import os
from PIL import Image, UnidentifiedImageError
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

class MoviePosterDataset(Dataset):
    def __init__(self, root_dir, movie_info_df, transform = None):
        self.root_dir = root_dir
        self.movie_info_df = movie_info_df
        self.transform = transform

    def __len__(self):
        return len(self.movie_info_df)

    def __getitem__(self, idx):

        img_path = os.path.join(self.root_dir, str(self.movie_info_df.iloc[idx]['id']) + '.jpg')
        # if not os.path.exists(img_path):
        #   # If the image file doesn't exist, skip the iteration
        #   return None, None, None, None, None, None, None, None
        # print(img_path)
        image = Image.open(img_path)


        if self.transform:
            image = self.transform(image)

        genre = self.movie_info_df.iloc[idx]['genre']
        movie_id = self.movie_info_df.iloc[idx]['id']
        date = self.movie_info_df.iloc[idx]['date']
        minute = self.movie_info_df.iloc[idx]['minute']
        rating = self.movie_info_df.iloc[idx]['rating']
        name = str(self.movie_info_df.iloc[idx]['name']) if not pd.isna(self.movie_info_df.iloc[idx]['name']) else ""
        description = str(self.movie_info_df.iloc[idx]['description']) if not pd.isna(self.movie_info_df.iloc[idx]['description']) else ""
        tagline = str(self.movie_info_df.iloc[idx]['tagline']) if not pd.isna(self.movie_info_df.iloc[idx]['tagline']) else ""
        # name = self.movie_info_df.iloc[idx]['name']
        # description = self.movie_info_df.iloc[idx]['description']
        # tagline = self.movie_info_df.iloc[idx]['tagline']
        return image, movie_id, date, minute, rating, name, description, tagline, genre
```

```python
import numpy as np
import timm.data.auto_augment as auto_augment

from sklearn.model_selection import train_test_split

# Assuming you have a DataFrame 'df' with the following columns:
# 'images', 'movie_ids', 'dates', 'ratings', 'languages', 'descriptions', 'taglines', 'labels'

data_transforms = transforms.Compose([
    # auto_augment.AutoAugment(policy = 'original'),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Split the data into training and validation sets
train_idx, val_idx = train_test_split(
    movie_info_df.index,
    test_size = 0.2,
    random_state = 42
)

# Create the training dataset
train_dataset = MoviePosterDataset(
    root_dir='/kaggle/input/letterboxd/posters',
    movie_info_df = movie_info_df,
    transform = data_transforms
)

# Create the validation dataset
val_dataset = MoviePosterDataset(
    root_dir='/kaggle/input/letterboxd/posters',
    movie_info_df = movie_info_df,
    transform= data_transforms
)

train_loader = DataLoader(train_dataset, batch_size= 1024, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size= 1024, shuffle=False, num_workers=4)
```

Efficient Net v2 Small

```python
from torchvision.models import efficientnet_v2_s, vit_b_16
import torch.nn as nn

# Get the total number of class in data
num_classes = len(movie_info_df['genre'].unique())

# Load the pre-trained EfficientNet-V2 model
model = efficientnet_v2_s(pretrained=True)
```

```python
# Select the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Move the entire model to the device
model.to(device)

# Freeze the feature extractor layers
for param in model.features.parameters():
    param.requires_grad = False

# Keep the last layer trainable
for param in model.classifier[-1].parameters():
    param.requires_grad = True

# Move the feature extractor layers to CPU
model.features.to("cpu")

# Get the number of features in the last layer
num_features = model.classifier[-1].in_features

# Replace the last layer with a new linear layer
model.classifier[-1] = nn.Linear(num_features, num_classes)
```

```python
import torch.nn.functional as F
from tqdm import tqdm

# Set the device (CPU or GPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Set up the optimizer and loss function
optimizer = torch.optim.Adam(model.classifier[-1].parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

# Training loop
num_epochs = 10
for epoch in range(num_epochs):
    train_loader_tqdm = tqdm(train_loader, desc=f'Epoch {epoch}', unit='batch')
    for images, movie_id, date, minute, rating, name, description, tagline, labels in train_loader_tqdm:
        images = images.to(device)
        labels = labels.to(device)

        # Turn it to one-hot vectors
        labels = F.one_hot(labels, num_classes=num_classes).float()

        # Forward pass
        outputs = model(images)
        outputs = F.softmax(outputs, dim=1)
        print()
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Update the progress bar
        train_loader_tqdm.set_postfix(loss=loss.item())
```

```python
from sklearn.metrics import roc_auc_score, f1_score, precision_score, accuracy_score

# Test the model
model.eval()
all_outputs = []
all_labels = []
with torch.no_grad():
    for images, movie_id, date, minute, rating, name, description, tagline, labels in tqdm(val_loader, desc='Testing', unit='batch'):
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        outputs = F.softmax(outputs, dim=1)

        all_outputs.extend(outputs.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Calculate the metrics
auc_roc = roc_auc_score(F.one_hot(torch.tensor(all_labels), num_classes=num_classes).float(), all_outputs)
f1 = f1_score(all_labels, np.argmax(all_outputs, axis=1), average='macro')
precision = precision_score(all_labels, np.argmax(all_outputs, axis=1), average='macro')
accuracy = accuracy_score(all_labels, np.argmax(all_outputs, axis=1))

print(f'AUC-ROC: {auc_roc:.4f}')
print(f'F1-score: {f1:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Accuracy: {accuracy:.4f}')
```
